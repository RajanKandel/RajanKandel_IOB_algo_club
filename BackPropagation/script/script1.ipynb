```python
import numpy as np 
```

```python
# generate random data points (n=5) in the circle  with randomly generated center and radius 
# save those set of point as a list of list (n = 1000) as a training example 
```

```python
def generate_circle_points(center_x, center_y, radius):
    """
    Generate 5 random points on the circumference of a circle with given center and radius.
    
    Parameters:
    center_x (float): X coordinate of circle center
    center_y (float): Y coordinate of circle center
    radius (float): Radius of the circle
    
    Returns:
    list: List of tuples containing (x, y) coordinates of generated points
    """
    num_points = 5
    
    # Generate random angles between 0 and 2Ï€
    theta = np.random.uniform(0, 2*np.pi, num_points)
    
    # Convert polar coordinates to Cartesian coordinates
    # Use exact radius to ensure points lie on circumference
    x = center_x + radius * np.cos(theta)
    y = center_y + radius * np.sin(theta)
    
    # Combine coordinates into points
    points = list(zip(x, y))
    
    return points
```

```python
#check whether all those points lie in a circumference
```

```python
center_x =4
center_y =5
radius= 6.5
points_in_circle = generate_circle_points(center_x, center_y, radius)

print("Generated points:")
for i, (x, y) in enumerate(points_in_circle, 1):
    print(f"Point {i}: ({x:.2f}, {y:.2f})")
```

```python

for x,y in points_in_circle:
        distance = np.sqrt((x - center_x)**2 + (y - center_y)**2)
        print(f"Distance from center: {distance:.2f} (should be = {radius})")
```

```python
#generate the training dataset 
```

```python
num_datset = 1000
x_centers =  np.random.uniform(0, 20,num_datset)
y_centers =  np.random.uniform(0, 20,num_datset)
radii =  np.random.uniform(0, 10, num_datset)
```

```python
training_data =  []
for i in range (0, num_datset):
    points_in_circle = generate_circle_points(x_centers[i], y_centers[i], radii[i])
    training_data.append(points_in_circle)
```

```python
print(f'#Training data set size : {len(training_data)}')
for data in training_data:
    print(data)
```
